# **Feature Selection**

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for col in app_df:
    if app_df[col].dtypes=='object':
        app_df[col] = LE.fit_transform(app_df[col].astype(str))
app_df.head()
```

```python
#print(app_df)
app_df.corr()
```

```python
features = app_df.drop(['CommName','Arrivals','YardName'], axis=1)
label = app_df['Model']
```

# **Machine Learning Model**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state = 10)
```

```python
if 'Date' in app_df.columns:
    app_df.drop('Date', axis=1, inplace=True)
else:
    print("Column 'Date' not found in dataframe.")

print(x_train, x_test, y_train, y_test)
```

```python

from xgboost import XGBRegressor

XGB_model = XGBRegressor()

XGB_model.fit(x_train, y_train)

print('XGBoost Model Accuracy : ', XGB_model.score(x_test, y_test)*100, '%')
```
